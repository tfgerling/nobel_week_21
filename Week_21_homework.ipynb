#### What types of APIs are there?

There are four general types of APIs used in web applications:
- Public APIs - open and available for use by any outside developer or business to enable sharing of applications.
- Partner APIs - offered to specific third-party business partner developers or API consumcers
- Composite APIs - combine two or more APIs for related operations to import speed and performance
- Internal APIs - internal enterprise used to connect systems and data with the business

Types of API Architectures:
- REST - key part of modern web application development. There are 6 rules that must be followed for an API to comply with REST (Representational State Transfer). 1. Cannot keep the information received from client/sender. 2. The client and server are independent of each other. Changes made on one doesn't affect the other. 3. Must have caching ability which allows information to be stored for a certain period in order to make the app more efficient and faster. 4. Have a uniform interface for communication. 5. Be a layered system, letting legacy code be encapsulated in one part of the API. 6. Adhere to code on demand allowing code to be transmitted within an application through the API.
- RPC - Remote Procedure Call - oldest and simplest type of API. The client executes coe on the server. This makes it easier for developers to create applications involving multiple programs or services.
- SOAP - Simple Object Access Protocol - offical protocol defined by the World Wide Web Consortium (W3C) and comes with strict rules. 

#### Why are APIs important?
APIs are a way for systems/applications to communicate and pass data. A well-written API can provide some very important long- term benefits, such as:
- Integration with other systems (we use APIs to access information from SonarQube so we can display it in dashboards)
- Take control - you can access information using the API and present it in a better format, for example
- Engage your customers and vendors
